```python
from dash import Dash, html, dcc, callback, Output, Input, dash_table
import pandas as pd
import plotly.express as px

# Load data
df = pd.read_excel('MP-MLA-DASH.xlsx')
df['Forward Year'] = df['Forward Year'].astype(str)

# Initialize Dash app
app = Dash(__name__)

# Define app layout
app.layout = html.Div([
    html.H1(
        children='MP_MLA_R&B Analysis Dashboard',
        style={
            'color': '#FFA823',
            'textAlign': 'center',
            'fontSize': '40px',
            'fontFamily': 'Arial, sans-serif',
            'fontWeight': 'bold',
            'backgroundColor': '#387478',
            'padding': '15px',
            'borderRadius': '10px',
            'boxShadow': '0px 2px 4px rgba(0, 0, 0, 0.1)'
        }
    ),
    html.Br(),
    html.Div([
        html.Label("Primary Filter"),
        dcc.Dropdown(
            id='primary-filter',
            options=[{'label': col, 'value': col} for col in df.columns],
            placeholder="Select a Primary Filter",
            style={'marginBottom': '10px'}
        ),
        html.Label("Secondary Filter"),
        dcc.Dropdown(
            id='secondary-filter',
            multi=True,
            placeholder="Select Secondary Filters",
            style={'marginBottom': '20px'}
        ),
    ]),
    html.Div(id='graphs-container')
])

# Callback to update secondary filter based on primary filter
@callback(
    Output('secondary-filter', 'options'),
    [Input('primary-filter', 'value')]
)
def update_filters(primary):
    if not primary:
        return []
    secondary_options = [{'label': val, 'value': val} for val in df[primary].dropna().unique()]
    return secondary_options

# Helper function to create styled tables with search and pagination
def create_styled_table(df, table_id):
    return dash_table.DataTable(
        id=table_id,
        columns=[{'name': col, 'id': col, 'presentation': 'markdown'} for col in df.columns],
        data=df.to_dict('records'),
        sort_action='native',
        filter_action='native',
        page_action='native',
        page_current=0,
        page_size=10,
        style_table={
            'border': '1px solid #003f5c',
            'width': '80%',
            'margin': 'auto',
        },
        style_cell={
            'padding': '5px',
            'textAlign': 'left',
            'border': '1px solid #003f5c',
            'backgroundColor': '#FAFAFA',
            'minWidth': '70px',
            'maxWidth': '200px',
            'whiteSpace': 'normal',
        },
        style_header={
            'backgroundColor': '#003f5c',
            'color': 'white',
            'fontWeight': 'bold',
            'textAlign': 'center',
            'border': '1px solid #003f5c'
        },
    )

# Callback to update graphs and tables based on selected filters
@callback(
    Output('graphs-container', 'children'),
    [Input('primary-filter', 'value'),
     Input('secondary-filter', 'value')]
)
def display_graphs(primary, secondary_values):
    filtered_df = df

    # Apply filters
    if primary and secondary_values:
        filtered_df = filtered_df[filtered_df[primary].isin(secondary_values)]

    sections = []

    # Summary Section
    filtered_df = filtered_df.dropna(subset=['L1 Category', 'Disposal Type'])

    # L1 Category Graph
    l1_category_df = filtered_df['L1 Category'].value_counts().reset_index()
    l1_category_df.columns = ['L1 Category', 'Count']
    
    l1_fig = px.bar(
        l1_category_df,
        x='L1 Category',
        y='Count',
        title="L1 Category",
        color_discrete_sequence=['#FFF574'],
        text_auto=True
    )    
    l1_fig.update_layout(
        height=700,
        width=700,
        title={
            'text': "L1 Category",
            'font': {'size': 22, 'weight': 'bold'}  # Increase the title size and make it bold
        },
        xaxis_title={
            'font': {'size': 12, 'weight': 'bold'}  # Make the x-axis label bold and larger
        },
        yaxis_title={
            'font': {'size': 18, 'weight': 'bold'}  # Make the y-axis label bold and larger
        },
        font={
            'size': 13,  # Set the default font size for the tick labels
            'weight': 'bold'  # Make the tick labels bold
        }    
    )
    
    # To make the text on bars bold and large
    l1_fig.update_traces(textfont=dict(size=14, family='Arial', weight='bold'))  # Bold and large text on bars


    # Disposal Type Graph
    disposal_type_df = filtered_df['Disposal Type'].value_counts().reset_index()
    disposal_type_df.columns = ['Disposal Type', 'Count']
    disposal_fig = px.bar(
        disposal_type_df,
        x='Count',
        y='Disposal Type',
        title="Disposal Type",
        orientation='h',
        color_discrete_sequence=['#FFF574'],
        text_auto=True
    )

    disposal_fig.update_layout(
    height=700,  # Increase height
    width=600,    # Increase width
    title={
            'text': "Disposal Type",
            'font': {'size': 22, 'weight': 'bold'}  # Increase the title size and make it bold
        },
        xaxis_title={
            'font': {'size': 12, 'weight': 'bold'}  # Make the x-axis label bold and larger
        },
        yaxis_title={
            'font': {'size': 18, 'weight': 'bold'}  # Make the y-axis label bold and larger
        },
        font={
            'size': 13,  # Set the default font size for the tick labels
            'weight': 'bold'  # Make the tick labels bold
        }    )

# To make the text on bars bold and large
    l1_fig.update_traces(textfont=dict(size=14, family='Arial', weight='bold'))        


    sections.append(html.Div([
        html.H1("Summary", style={'color': '#003f5c', 'textAlign': 'center'}),
        html.Div([
            html.Div([
                html.H2("Total References", style={'textAlign': 'center', 'color': '#003f5c'}),
                html.H2(filtered_df['MPMLA_LetterID'].count(), style={'textAlign': 'center', 'color': '#8000ff'})
            ], style={'width': '30%', 'display': 'inline-block', 'background': '#D3F1DF', 'border': '1px solid #003f5c', 'padding': '5px', 'borderRadius': '3px', 'margin': '5px', 'boxShadow': '0px 2px 5px rgba(0,0,0,0.1)'}),
            html.Div([
                html.H2("Average Time of Disposal (Days)", style={'textAlign': 'center', 'color': '#003f5c'}),
                html.H2(f"{filtered_df['Days Taken To Close The Application'].mean():.2f}", style={'textAlign': 'center', 'color': '#8000ff'})
            ], style={'width': '30%', 'display': 'inline-block', 'background': '#D3F1DF', 'border': '1px solid #003f5c', 'padding': '5px', 'borderRadius': '3px', 'margin': '5px', 'boxShadow': '0px 2px 5px rgba(0,0,0,0.1)'}),
        ], style={'display': 'flex', 'justifyContent': 'center'}),

        html.Br(),
        html.Div([
            html.Div([dcc.Graph(figure=l1_fig)], style={'width': '48%', 'display': 'inline-block'}),
            html.Div([dcc.Graph(figure=disposal_fig)], style={'width': '48%', 'display': 'inline-block'})
        ], style={'display': 'flex', 'justifyContent': 'space-between'})
    ]))

    # Level-wise Categorization Section
    filtered_df = filtered_df.dropna(subset=['Applicant District', 'Forward Year', 'L2 Category', 'L3 Category'])

    # Applicant District Graph
    applicant_district_df = filtered_df['Applicant District'].value_counts().reset_index()
    applicant_district_df.columns = ['Applicant District', 'Count']
    applicant_district_fig = px.bar(
        applicant_district_df,
        x='Count',
        y='Applicant District',
        title="Applicant District and their Count",
        color_discrete_sequence=['#003f5c'],
        text_auto=True
    )
    applicant_district_fig.update_layout(
      height=900,
      width=800
    )

    # Forward Year Graph
    forward_year_df = filtered_df['Forward Year'].value_counts().reset_index()
    forward_year_df.columns = ['Forward Year', 'Count']
    forward_year_fig = px.bar(
        forward_year_df,
        x='Forward Year',
        y='Count',
        title="Forward Year and their Count",
        color_discrete_sequence=['#003f5c'],
        text_auto=True
    )
    forward_year_fig.update_layout(
      height=900,
      width=600
    )

    l2_category_df = filtered_df.groupby('L2 Category').size().reset_index(name='Count')
    l3_category_df = filtered_df.groupby('L3 Category').size().reset_index(name='Count')

    sections.append(html.Div([
        html.H2("Level-wise Categorization", style={'color': '#003f5c', 'textAlign': 'center'}),
        html.Div([
            html.Div([dcc.Graph(figure=applicant_district_fig)], style={'width': '48%', 'display': 'inline-block'}),
            html.Div([dcc.Graph(figure=forward_year_fig)], style={'width': '48%', 'display': 'inline-block'})
        ], style={'display': 'flex', 'justifyContent': 'space-between'}),
        html.Div([
            html.Div([
                html.H4("L2 Category and Count"),
                create_styled_table(l2_category_df, 'l2-category-table')
            ], style={'width': '48%', 'display': 'inline-block'}),
            html.Div([
                html.H4("L3 Category and Count"),
                create_styled_table(l3_category_df, 'l3-category-table')
            ], style={'width': '48%', 'display': 'inline-block'})
        ])
    ]))

    # MP_MLA-wise Analysis Section
    filtered_df = filtered_df.dropna(subset=['MP_MLA_Details', 'Timely Disposed Green Channel', 'Timely Disposed Yellow Channel'])

    mla_details_df = filtered_df.groupby(['MP_MLA_Details','Assembly_Name']).size().reset_index(name='Count')

        # Donut Charts
    green_channel_df = filtered_df.groupby('Timely Disposed Green Channel').size().reset_index(name='Count')
    green_channel_fig = px.pie(
        green_channel_df,
        names='Timely Disposed Green Channel',
        values='Count',
        title="Timely Disposed Green Channel",
        hole=0.4,
        color_discrete_map={'Yes': 'green', 'No': '#ffa600'}  # Change "Yes" to Green, "No" to Yellow
    )
    
#     Donut Charts
    green_channel_df = filtered_df.groupby('Timely Disposed Green Channel').size().reset_index(name='Count')
    green_channel_fig = px.pie(
        green_channel_df,
        names='Timely Disposed Green Channel',
        values='Count',
        title="Timely Disposed Green Channel",
        hole=0.4,
        color_discrete_map={'Yes': 'green', 'No': 'red'}
    )
    
    # Update title styling for Green Channel
    green_channel_fig.update_layout(
        title={'text': "Timely Disposed Green Channel", 'font': {'size': 20, 'color': '#002E00', 'family': 'Arial', 'weight': 'bold'}},
        colorway=['green', 'red']  # Define colorway for the chart
    )
    
    yellow_channel_df = filtered_df.groupby('Timely Disposed Yellow Channel').size().reset_index(name='Count')
    yellow_channel_fig = px.pie(
        yellow_channel_df,
        names='Timely Disposed Yellow Channel',
        values='Count',
        title="Timely Disposed Yellow Channel",
        hole=0.4,
        color_discrete_map={'Yes': 'yellow', 'No': 'red'}
    )
    
    # Update title styling for Yellow Channel
    yellow_channel_fig.update_layout(
        title={'text': "Timely Disposed Yellow Channel", 'font': {'size': 20, 'color': '#FFCC00', 'family': 'Arial', 'weight': 'bold'}},
        colorway=['yellow', 'red']  # Define colorway for the chart
    )

    sections.append(html.Div([
        html.H2("MP_MLA-wise Analysis", style={'color': '#003f5c', 'textAlign': 'center'}),
        html.Div([
            html.Div([
                html.H4("MP_MLA_Details and Count"),
                create_styled_table(mla_details_df, 'mla-details-table')
            ], style={'width': '48%', 'display': 'inline-block'}),
            html.Div([
                dcc.Graph(figure=green_channel_fig),
                dcc.Graph(figure=yellow_channel_fig)
            ], style={'width': '48%', 'display': 'inline-block'})
        ])
    ]))

    # Repeated Subject & Answers Section
    filtered_df = filtered_df.dropna(subset=['Subject', 'Answer_Brief'])

    subjects_df = filtered_df.groupby('Subject').size().reset_index(name='Count')
    answer_brief_df = filtered_df.groupby('Answer_Brief').size().reset_index(name='Count')

    sections.append(html.Div([
        html.H2("Repeated Subject & Answers", style={'color': '#003f5c', 'textAlign': 'center'}),
        html.Div([
            html.Div([
                html.H4("Subjects and Count"),
                create_styled_table(subjects_df, 'subjects-table')
            ], style={'width': '48%', 'display': 'inline-block'}),
            html.Div([
                html.H4("Answer Brief and Count"),
                create_styled_table(answer_brief_df, 'answer-brief-table')
            ], style={'width': '48%', 'display': 'inline-block'})
        ])
    ]))

    return html.Div(sections)

if __name__ == '__main__':
    app.run_server(debug=True)
```
